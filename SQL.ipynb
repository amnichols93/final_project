```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
from config import db_password
```

```python
allteam_df = pd.read_csv('Resources/allteam_csvs/allteam_df_updated.csv')
```

```python
unique_teams = allteam_df['Base_Team'].unique()
```

```python
column_names = ["ID", "Team"]
team_df = pd.DataFrame(columns = column_names)
```

```python
team_df['ID'] = range(1,33,1)
team_df['Team'] = unique_teams
team_df
```

```python
coded_df = allteam_df.copy()
```

```python
for team in unique_teams:
    id_df = team_df[team_df['Team'] == team]
    team_id = id_df['ID'].sum()
    coded_df['Base_Team'] = coded_df['Base_Team'].replace(team, team_id)
    coded_df['Opp'] = coded_df['Opp'].replace(team, team_id)
```

```python
coded_df
```

```python
# Get coded_df and team_df into SQL

db_string = f"postgres://postgres:{db_password}@127.0.0.1:5432/PickSys"
engine = create_engine(db_string)
coded_df.to_sql(name='coded_stats', con=engine, if_exists='append')
team_df.to_sql(name='team_IDs', con=engine, if_exists='append')
```
